필요한 라이브러리를 가져옵니다.

```python
from bs4 import BeautifulSoup
from html_table_parser import parser_functions as parser
import pandas as pd
import requests
```

크롤링할 주소를 url에 저장한다. 아라시의 음반 목록을 가져온다.

```python
# Crawling
url = 'https://en.wikipedia.org/wiki/Arashi_discography'
```

해당 url에 대하여 HTTP request를 통해 HTML response를 받는다.

```python
req = requests.get(url)
html = req.text
soup = BeautifulSoup(html, 'html.parser')
```

해당 웹페이지에서 크롤링하고자 하는 표의 속성을 확인 한후 fina_all() 함수를 통해 HTML에서 해당 표만을 추출합니다. 아래의 경우 HTML에서 class가 sortable wikitable인 table 태그만을 추출하고자 한다.

```python
# 크롤링하고자 하는 표 : <tabl class="wikitable">
tables = soup.find_all('table', attrs={'class' : 'wikitable'})
```

변수 tables에 담겨있는 정보들을 데이터프레임 형태로 변환한다.

```python
# 데이터 프레임 설정
temp = []
for table in tables :
  table = parser.make2d(table)
  table = pd.DataFrame(table[1:])
  temp.append(table)
df = pd.concat(temp)
df
```

데이터프레임의 간략한 정보를 확인한다.

```python
df.info()
```

크롤링 후 후처리를 한다.

```python
df.columns = ['Title', 'Album details', 'JPN', 'JPNHot',	'TWN', 'TWNEA',	'KOR',	'KORINT',	'Certifications',	'First week',	'Total',	'Album'] # 컬럼명 변경
df.dropna(inplace=True) # 결측치 제거
df.reset_index(drop=True, inplace=True) # 인덱스 재설정
df = df.drop([0], axis=0) # 맨위 행 title 제거
df = df.drop([33],axis = 0)
```
