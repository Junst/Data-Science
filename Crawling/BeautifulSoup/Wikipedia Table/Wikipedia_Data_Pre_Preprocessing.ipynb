저번에 저장해놓은 데이터 셋을 불러온다.

```python
df = pd.read_csv('WikipediaTable.csv', index_col=0)
df
```

데이터프레임의 간략한 정보를 확인

```python
df.info()
```

###Title 전처리
데이터셋의 Title 컬럼에 대한 전처리 함수를 선언

```python
def PreprocessTitle(string):
  import re
  p = re.compile('[^""]+')
  return p.search(string).group()
```

```python
df['Title'] = df['Title'].apply(PreprocessTitle)
```

```python
df
```

## Ranking 전처리
데이터셋의 Ranking 컬럼 drop 

```python
df = df.drop(['JPN'],axis = 1)
df = df.drop(['JPNHot'],axis = 1)
df = df.drop(['TWN'],axis = 1)
df = df.drop(['TWNEA'],axis = 1)
df = df.drop(['KOR'],axis = 1)
df = df.drop(['KORINT'],axis = 1)
```

```python
df
```

## Certifications 전처리
주석 제거



```python
df['Certifications'] = df['Certifications'].str.replace(pat=r'[',repl=r'',regex=True)
df['Certifications'] = df['Certifications'].str.replace(pat=r']',repl=r'',regex=True)
df['Certifications'] = df['Certifications'].replace('RIAJ: Platinum23','RIAJ: Platinum')
df['Certifications'] = df['Certifications'].replace('RIAJ: 2× Platinum23','RIAJ: 2× Platinum')
df['Certifications']= df['Certifications'].replace('RIAJ: 3× Platinum23','RIAJ: 3× Platinum')
df['Certifications'] = df['Certifications'].replace('RIAJ: Million23','RIAJ: Million')
```

```python
df
```

```python
df.to_csv('WikipediaPreprocessingTable.csv')
```
